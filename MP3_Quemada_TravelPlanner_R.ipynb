Travel Planner




```python
import readline

# Prompt the user for travel preferences and store the input
trip_type = input("Enter your preferred trip type (relaxed, adventurous): ")
readline.get_line_buffer()

budget = input("Enter your preferred budget (low, moderate, high): ")
readline.get_line_buffer()

travel_companions = input("Enter your travel companions (solo, family, friends): ")
readline.get_line_buffer()

duration = input("Enter your preferred duration (short, moderate, long): ")
readline.get_line_buffer()

# Define and apply rules based on user input

# Rule 1: Trip Type and Budget
if trip_type == "relaxed" and budget == "high":
    print("\nRule Applied: Relaxed trip with high budget.")
    print("Conclusion: Consider a luxury resort stay.")
elif trip_type == "relaxed" and budget == "moderate":
    print("\nRule Applied: Relaxed trip with moderate budget.")
    print("Conclusion: Look for comfortable hotels or vacation rentals.")
elif trip_type == "relaxed" and budget == "low":
    print("\nRule Applied: Relaxed trip with low budget.")
    print("Conclusion: Explore budget-friendly guesthouses or camping options.")
elif trip_type == "adventurous" and budget == "high":
    print("\nRule Applied: Adventurous trip with high budget.")
    print("Conclusion: Consider an exotic adventure.")
elif trip_type == "adventurous" and budget == "moderate":
    print("\nRule Applied: Adventurous trip with moderate budget.")
    print("Conclusion: Look for adventure packages or guided tours.")
elif trip_type == "adventurous" and budget == "low":
    print("\nRule Applied: Adventurous trip with low budget.")
    print("Conclusion: Explore hiking, camping, or budget adventure activities.")
else:
    print("\nRule Applied: Invalid trip type or budget input in Rule 1.")
    print("Conclusion: Please enter valid options for trip type and budget.")


# Rule 2: Trip Type and Travel Companions
# This rule will now be applied after the first rule and can provide additional conclusions
if trip_type == "adventurous" and travel_companions == "solo":
    print("\nRule Applied: Adventurous trip with solo travel.")
    print("Conclusion: Consider backpacking or a solo adventure tour.")
elif trip_type == "adventurous" and travel_companions == "family":
    print("\nRule Applied: Adventurous trip with family.")
    print("Conclusion: Look for family-friendly adventure activities or destinations.")
elif travel_companions not in ["solo", "family", "friends"]:
     print("\nRule Applied: Invalid travel companions input in Rule 2.")
     print("Conclusion: Please enter either 'solo', 'family', or 'friends' for travel companions.")


# Rule 3: Duration and Budget (Chained Rule)
recommended_activities = []
if duration == "long" and budget == "high":
    print("\nRule Applied: Long duration with high budget.")
    print("Intermediate Conclusion: Suggest exploring multiple destinations or extended stays.")
    # Chained rule based on the intermediate conclusion
    if trip_type == "adventurous":
        print("Chained Rule Applied: Adventurous trip with long duration and high budget.")
        print("Conclusion: Consider a multi-country adventure expedition.")
        recommended_activities.append("multi-country expedition")
    else:
        print("Chained Rule Applied: Relaxed trip with long duration and high budget.")
        print("Conclusion: Consider a long-term stay in a luxury villa.")
        recommended_activities.append("luxury villa stay")
elif duration == "long" and budget == "moderate":
    print("\nRule Applied: Long duration with moderate budget.")
    print("Conclusion: Explore extended stays in affordable accommodations.")
    recommended_activities.append("extended affordable stay")
elif duration not in ["short", "moderate", "long"]:
    print("\nRule Applied: Invalid duration input in Rule 3.")
    print("Conclusion: Please enter either 'short', 'moderate', or 'long' for duration.")
elif budget not in ["low", "moderate", "high"]:
     print("\nRule Applied: Invalid budget input in Rule 3.")
     print("Conclusion: Please enter either 'low', 'moderate', or 'high' for budget.")


# Rule 4: Travel Companions and Duration
if travel_companions == "friends" and duration == "short":
    print("\nRule Applied: Friends travelling for a short duration.")
    print("Conclusion: Suggest a city break or a nearby getaway.")

# Rule 5: Budget and Duration
if budget == "low" and duration == "short":
    print("\nRule Applied: Low budget and short duration.")
    print("Conclusion: Recommend local attractions or a road trip to a nearby destination.")

# Display recommended activities if any
if recommended_activities:
    print(f"\nBased on your preferences, here are some recommended activities: {', '.join(recommended_activities)}")
```
